```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from faker import Faker
```

```python
fake = Faker()

def simulated_data(num_transcations):
  transaction=[]

  for i in range(num_transactions):
    transaction_date = fake.date_time_between(start_date ='-365d',end_date = 'now')
    transaction_id = fake.uuid4()
    merchant= fake.company()
    name = fake.name()
    country = fake.country()
    amount = np.round(np.random.uniform(100,1000),2)

    category = np.random.choice(['Entertainment','Retail','Electronics','Grocery','Education','Travel'])
    card_category = np.random.choice(['Business','Platinum','Gold'])
    customer_id = fake.random_number(digits=5)

    transaction.append({'Transaction Date': transaction_date,
                        'Transaction ID':transaction_id,
                        'Company':merchant,
                        'Name':name,
                        'Country':country,
                        'Amount':amount,

                        'Category':category,
                        'Card Type':card_category,
                        'Customer ID':customer_id})

  return transaction
```

```python
num_transactions = 1000

data_1 = simulated_data(num_transactions)
data = pd.DataFrame(data_1)

data.head(20)
```
